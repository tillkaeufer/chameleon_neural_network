# Plot Training

```python
def plot_history(history,log=False):
    loss_list = [s for s in history.keys() if 'loss' in s and 'val' not in s]
    val_loss_list = [s for s in history.keys() if 'loss' in s and 'val' in s]  
    if len(loss_list) == 0:
        print('Loss is missing in history')
        return 
    ## As loss always exists
    epochs = range(1,len(history[loss_list[0]]) + 1)
    
    ## Loss
    plt.figure(figsize=(9,6))
    for l in loss_list:
        plt.plot(epochs, history[l], 'tab:blue', label=f'Training loss: %5.3e' %history[l][-1],
                linestyle='none',marker='+')
    for l in val_loss_list:
        plt.plot(epochs, history[l], 'tab:orange', label='Test loss: %5.3e' %history[l][-1],
                linestyle='none',marker='+')
    
    plt.title('Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    if log:
        plt.yscale('log')
    plt.legend()
    plt.show()
    
```

```python
plot_history(history=history.history,log=False)
```
